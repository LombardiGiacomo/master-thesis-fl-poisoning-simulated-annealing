```python
import matplotlib.pyplot as plt
import numpy as np

# Dati
k_values = [0.5, 1.0, 2.0, 5.0, 10.0, 15.0, 20.0]

prec_km = [0.0811, 0.3333, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000]
rec_km = [0.1500, 0.5000, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000]
f1_km = [0.1053, 0.4000, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000]

prec_dist = [0.2727, 0.9091, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000]
rec_dist = [0.3000, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000]
f1_dist = [0.2857, 0.9524, 1.0000, 1.0000, 1.0000, 1.0000, 1.0000]

labels = ['Precision', 'Recall', 'F-Measure']
x = np.arange(len(labels))
width = 0.35

# Impostiamo una griglia 2x4 per ospitare 7 grafici (2 righe, 4 colonne)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten() # Appiattiamo l'array 2D in 1D per iterarci facilmente

fig.suptitle("Security metrices comparison at round 20 for different noise magnitudes (k)", fontsize=20, y=1.02, fontweight='bold')

for i, k in enumerate(k_values):
    ax = axes[i] # Seleziona il subplot corrente
    
    # Estraiamo i punteggi per il k corrente (indice i)
    kmeans_scores = [prec_km[i], rec_km[i], f1_km[i]]
    dist_scores = [prec_dist[i], rec_dist[i], f1_dist[i]]
    
    # Disegno delle barre
    rects1 = ax.bar(x - width/2, kmeans_scores, width, label='K-Means', color='#d62728', alpha=0.9)
    rects2 = ax.bar(x + width/2, dist_scores, width, label='Distance', color='#1f77b4', alpha=0.9)
    
    # Titolo e formattazione assi per il singolo riquadro
    ax.set_title(f"Attack with k = {k}", fontsize=15)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=13)
    ax.set_ylim(0, 1.15)
    
    # Inseriamo la legenda SOLO nel primissimo grafico, per non sporcare il disegno
    if i == 0:
        ax.legend(fontsize=12, loc='upper left')
        
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Scriviamo i valori numerici sopra ogni singola barra
    for rect in rects1 + rects2:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), # Piccola spaziatura dal bordo superiore della barra
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

# Nascondiamo l'ottavo subplot vuoto (indice 7)
axes[7].set_visible(False)

plt.tight_layout()
plt.savefig("./GraphsImages/All_k_metrices_comparison_Bar_Charts.png", dpi=300, bbox_inches='tight')
plt.show()
```
